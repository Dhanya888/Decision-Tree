```python
from sklearn.preprocessing import LabelEncoder
raw=LabelEncoder()
df['GENDER']=raw.fit_transform(df['GENDER'])
```

```python
from sklearn.preprocessing import LabelEncoder
raw=LabelEncoder()
df['LUNG_CANCER']=raw.fit_transform(df['LUNG_CANCER'])
```

```python
z=df.corr()
sns.heatmap(z,annot=True)
```

```python
x=df.drop(['LUNG_CANCER'],axis=1)
y=df['LUNG_CANCER']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=95)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
wet=DecisionTreeClassifier(criterion='entropy')
wet.fit(x_train,y_train)
```

```python
y_pred=wet.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
from sklearn.svm import SVC
ter=SVC(kernel='linear')
ter.fit(x_train,y_train)
```

```python
y_pred=ter.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
yet=KNeighborsClassifier(n_neighbors=17)
yet.fit(x_train,y_train)
```

```python
y_pred=yet.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
sat=LogisticRegression()
sat.fit(x_train,y_train)
```

```python
y_pred=sat.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
qwe=RandomForestClassifier(n_estimators=100,n_jobs=1,random_state=70,max_depth=15,max_features=16,criterion='entropy',oob_score=True)
qwe.fit(x_train,y_train)
```

```python
y_pred=qwe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
from sklearn.naive_bayes import GaussianNB
red=GaussianNB()
red.fit(x_train,y_train)
```

```python
y_pred=red.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```
